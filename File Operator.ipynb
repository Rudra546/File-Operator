```python
from datetime import datetime

print("=== Simple Journal App ===")

file_name = "journal.txt"

while True:
    print("\n1. Add Entry")
    print("2. View Entries")
    print("3. Search Entry")
    print("4. Delete All")
    print("5. Exit")

    choice = input("Enter choice: ")


    if choice == "1":
        try:
            text = input("Write something: ")
            time = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

            f = open(file_name, "a")
            f.write(f"[{time}] {text}\n")
            f.close()

            print("Entry added successfully!")
        except Exception as e:
            print("Error while saving entry:", e)


    elif choice == "2":
        try:
            f = open(file_name, "r")
            data = f.readlines()
            f.close()

            if data:
                print("\nYour Entries:")
                for i in range(len(data)):
                    print(i+1, ".", data[i].strip())
            else:
                print("No entries found")

        except FileNotFoundError:
            print("No file found, add entry first")
        except Exception as e:
            print("Error while reading:", e)

    
    elif choice == "3":
        try:
            word = input("Enter keyword: ").lower()

            f = open(file_name, "r")
            data = f.readlines()
            f.close()

            found = False
            for line in data:
                if word in line.lower():
                    print(line.strip())
                    found = True

            if not found:
                print("No matching entry found")

        except FileNotFoundError:
            print("No file found")
        except Exception as e:
            print("Error while searching:", e)

    
    elif choice == "4":
        try:
            confirm = input("Delete all entries? (yes/no): ").lower()
            if confirm == "yes":
                f = open(file_name, "w")
                f.close()
                print("All entries deleted")
            else:
                print("Cancelled")

        except Exception as e:
            print("Error while deleting:", e)

    
    elif choice == "5":
        print("Thank you! Goodbye.")
        break

    else:
        print("Invalid choice")
```
